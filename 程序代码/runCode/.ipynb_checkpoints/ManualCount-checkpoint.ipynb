```python
import os
import pandas as pd
```

```python
rootDir = '/home/yfliu/manual_inspection/'
conPathList = ['after/segment_0.7.csv','after/segment_0.8.csv','after/segment_0.9.csv']
funPathList = ['after_function/segment_0.8.csv','after_function/segment_0.8.csv']
code_contract = pd.read_csv('../16w_contract_v2.csv')
snippet_contract = pd.read_csv('../snippet_contract_v2.csv').drop_duplicates(['dealed_code'])
```

```python
def getCount(pathList, rootDir):
    res = None
    for name in pathList:
        path = os.path.join(rootDir, name)
        manualCsv = pd.read_csv(path)
        filterCsv = manualCsv[manualCsv['check'].apply(lambda x : x <= 2.0)]
        if type(res) != type(filterCsv):
            #print('no')
            res = filterCsv
        else:
            #print('yes')
            res = res.append(filterCsv)
    return res
#con
conRes = getCount(conPathList, rootDir)
codeCode = code_contract['code']
idxList = []
for code in list(conRes['code']):
    idx = codeCode[codeCode == code].index
    if len(idx) > 1:
        print(idx)
    idxList.append(idx[0])
conIdxList = list(set(idxList))
#fun
funRes = getCount(funPathList, rootDir)
funIdxList = list(funRes.drop_duplicates(['code_index'])['code_index'])
#total
resList = list(set(conIdxList + funIdxList))
resList
```

```python
pd.DataFrame({'manualCount':resList}).to_csv('~/paper_data/manualCount.csv')
```
